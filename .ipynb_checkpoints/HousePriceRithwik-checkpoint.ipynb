```python
df2 = df.drop('society', axis='columns')
df2.shape
```

```python
df2['balcony'] = df2['balcony'].fillna(df2['balcony'].mean())
```

```python
df3 = df2.dropna()
df3.shape
```

```python
df3.isna().sum()
```

```python
df3.head()
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df3['total_sqft'].value_counts()
```

```python
total_sqft_int = []
for str_val in df3['total_sqft']:
    try:
        total_sqft_int.append(float(str_val))
    except:
        try:
            temp = []
            temp = str_val.split('-')
            total_sqft_int.append((float(temp[0])+float(temp[-1]))/2)
        except:
            total_sqft_int.append(np.nan)
```

```python
df4 = df3.reset_index(drop=True)
```

```python
df5 = df4.join(pd.DataFrame({'total_sqft_int':total_sqft_int}))
df5.head()
```

```python
df5.tail()
```

```python
df5.isna().sum()
```

```python
df6 = df5.dropna()
df6.shape
```

```python
df6.info()
```

```python
df6['size'].value_counts()
```

```python
size_int = []
for str_val in df6['size']:
    temp = []
    temp = str_val.split(" ")
    try:
        size_int.append(int(temp[0]))
    except:
        size_int.append(np.nan)
        print("Noise = ", str_val)
```

```python
df6 = df6.reset_index(drop=True)
```

```python
df7 = df6.join(pd.DataFrame({'bhk':size_int}))
df7.shape
```

```python
df7[df7['total_sqft_int']/df7['bhk'] < 350].head()
```

```python
df8 = df7[~(df7['total_sqft_int']/df7['bhk'] < 350)]
df8.shape
```

```python
df8['price_per_sqft'] = df8['price']*100000 / df8['total_sqft_int']
df8.head()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df9 = remove_pps_outliers(df8)
df9.shape
```

```python
df9.bath.unique()
```

```python
df9[df9.bath > df9.bhk+2]
```

```python
df10 = df9[df9.bath < df9.bhk+2]
df10.shape
```

```python
df11 = df10.drop(['area_type', 'availability', 'location', 'size', 'total_sqft'], axis=1)
df11.head()
```

```python
df11.to_csv("clean_data.csv", index=False)
```

```python
dff = pd.read_csv('clean_data.csv')
dff.shape
```

```python
X = dff.drop('price', axis=1)
y = dff['price']
print('shape of X = ', X.shape)
print('shape of y = ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 51)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr = LinearRegression()
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
lr.fit(X_train, y_train)
lr_score = lr.score(X_test, y_test)
lr_rmse = rmse(y_test, lr.predict(X_test))
lr_score, lr_rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
rfr_score = rfr.score(X_test, y_test)
rfr_rmse = rmse(y_test, rfr.predict(X_test))
rfr_score, rfr_rmse
```

```python
print(pd.DataFrame([{'Model': 'Linear Regression', 'Score': lr_score},
                   {'Model': 'RandomForest Regression', 'Score': rfr_score}]))
```
